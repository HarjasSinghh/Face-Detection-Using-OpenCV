```python
import numpy as np
import cv2 
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
test_image = cv2.imread('baby1.jpg')
test_image_gray = cv2.cvtColor(test_image, cv2.COLOR_BGR2GRAY)
plt.imshow(test_image_gray, cmap='gray')
```

```python
def convertToRGB(image):
    return cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

```python
haar_cascade_face = cv2.CascadeClassifier('haarcascade_frontalface_alt.xml')
```

```python
faces_rects = haar_cascade_face.detectMultiScale(test_image_gray, scaleFactor = 1.1, minNeighbors = 5);
print('Faces found: ', len(faces_rects))
```

```python
for (x,y,w,h) in faces_rects:
     cv2.rectangle(test_image, (x, y), (x+w, y+h), (0, 255, 0), 5)
        
```

```python
plt.imshow(convertToRGB(test_image))
```

```python
def detect_faces(cascade, test_image, scaleFactor = 1.09):
    image_copy = test_image.copy()
    gray_image = cv2.cvtColor(image_copy, cv2.COLOR_BGR2GRAY)
    faces_rect = cascade.detectMultiScale(gray_image, scaleFactor=scaleFactor, minNeighbors=2)
    for (x, y, w, h) in faces_rect:
        cv2.rectangle(image_copy, (x, y), (x+w, y+h), (0, 255, 0), 2)
        
    return image_copy
```

```python
test_image2 = cv2.imread('ID.jpeg')
faces = detect_faces(haar_cascade_face, test_image2)
plt.imshow(convertToRGB(faces))
```

```python
test_image2 = cv2.imread('group2.jpg')
faces = detect_faces(haar_cascade_face, test_image2)
plt.imshow(convertToRGB(faces))
```
